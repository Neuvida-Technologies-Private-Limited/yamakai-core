```python
from transformers import pipeline,GPT2Tokenizer,OPTForCausalLM
model=OPTForCausalLM.from_pretrained('/content/drive/MyDrive/yamakai-core/OPT_Paraphraser/pushkar_OPT_paraphaser')
tokenizer=GPT2Tokenizer.from_pretrained('/content/drive/MyDrive/yamakai-core/OPT_Paraphraser/pushkar_OPT_paraphaser',truncation=True)
```

```python
generator=pipeline("text-generation",model=model,tokenizer=tokenizer,device=0)
```

```python
def cleaned_para(input_sentence):
  p=generator('<s>'+input_sentence+ '</s>>>>><p>',do_sample=True,max_length=len(input_sentence.split(" "))+200,temperature = 0.8,repetition_penalty=1.2,top_p=0.4,top_k=1)
  return p[0]['generated_text'].split('</s>>>>><p>')[1].split('</p>')[0]
```

```python
from __future__ import unicode_literals, print_function
from spacy.lang.en import English # updated

def sentensizer(raw_text):
  nlp = English()
  nlp.add_pipe("sentencizer") # updated
  doc = nlp(raw_text)
  sentences = [sent for sent in doc.sents]
  print(sentences)
  return sentences
```

```python
def paraphraser(text):
  begin=time.time()
  x=[cleaned_para(str(i)) for i in sentensizer(text)]
  end=time.time()
  return (".".join(x),end-begin,len(".".join(x)))
```

```python
para="We encountered significant amounts of hardware misbehaviours in our compute cluster while teaching 'OPT-175B' The incident resulted in 40 manual changes during his 2 month cycle. During routine restarted training runs, the training interval was paused; a series of diagnostic tests were carried out to check if there were any issues with the node. Then, each node'cordoned-off status was tagged and training commenced from the very first save checkpoint. Given the fact that the number of hosts cycled out was larger than the number of manual restartes, we can confidently say that 80% of them were due to hardware failure"
print(para,len(para))
paraphraser(para)
```
